```python
# Criando um dataset
data = [['Fulano', 12, 7.0, True],
        ['Sicrano', 15, 3.5, False], 
        ['Beltrano', 18, 9.3, True]]

# Colocando nome nas colunas
# pd.DataFrame é para selecionar, filtrar, adicionar ou remove, cálcular, agrupar  e  funções

dados = pd.DataFrame(data, 
        columns = ['Aluno', 'Idade', 'Nota', 'Aprovado'])
dados
```

## Criando nova coluna e somando

```python
data = {'Idade': [25, 30, 35, 40],
        'nota': [80, 90, 85, 95]}

dados = pd.DataFrame(data)
```

```python
dados['testando'] = dados['Idade'] + dados['nota']
```

## Praticando

```python
# Criando um DataFrame a partir de um dicionário
data = {'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
        'Idade': [25, 30, 35, 40],
        'Cidade': ['São Paulo', 'Rio de Janeiro', 'Brasília', 'Salvador']}
df = pd.DataFrame(data)
```

```python
# Imprimindo o DataFrame
print("DataFrame:")
df
```

```python
# Acessando colunas
print("Acessando colunas:")
print(df['Nome'])            # Acessando uma coluna específica usando colchetes
print(df.Nome)               # Acessando uma coluna específica usando notação de ponto
```

```python
# Filtrando o DataFrame
print("Filtrando o DataFrame:")
filtro = df['Idade'] > 30
df[filtro]
```

```python
# Adicionando uma nova coluna
df['Profissão'] = ['Engenheiro', 'Advogada', 'Médico', 'Professor']
print("DataFrame com nova coluna:")
df
```

```python
# Salvando o DataFrame em um arquivo CSV
df.to_csv('dados.csv', index=False)
```

```python
# Lendo dados de um arquivo CSV
df_novo = pd.read_csv('dados.csv')
print("Novo DataFrame lido do arquivo CSV:")
df_novo
```
